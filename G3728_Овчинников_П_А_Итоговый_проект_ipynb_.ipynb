```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst) # shuffle() модуля random, перемешивает список
data = pd.DataFrame({'whoAmI':lst}) # сделали из списка DataFrame
# data.head()
print(data)

print("Вариант 1 Решение через pd.get_dummies")
# https://www.codecamp.ru/blog/pandas-get-dummies/
# pd.get_dummies(data, columns=['whoAmI'], drop_first=False)
print(pd.get_dummies(data, columns=['whoAmI'], drop_first=False))
# print(data) #т.е. data остался без изменений

print("Вариант 2 Решение без pd.get_dummies")
data_v2 = data.copy() #т.к. data в методе изменяется
data_v2['tmp'] = 1
data_v2.set_index([data_v2.index, 'whoAmI'], inplace=True)
data_v2 = data_v2.unstack(level=-1, fill_value = 0).astype(int)
data_v2.columns = data_v2.columns.droplevel()
data_v2.columns.name = None
print(data_v2)

print("Вариант 3 Решение без pd.get_dummies в одну строку")
# 4 функции для изменения форм таблиц в Pandas https://python-school.ru/blog/pandas-reshaping/ -  pivot stack unstack melt()  
data = data.melt(ignore_index=False).reset_index().pivot_table('variable', 'index', 'value', aggfunc='count').fillna(0).astype(int)
print(data)

```
